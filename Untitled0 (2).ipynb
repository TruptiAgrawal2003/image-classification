```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
data_train_path="/content/drive/MyDrive/Training_set"
```

```python
data_test_path="/content/drive/MyDrive/test"
```

```python
# data_val_path="/content/drive/MyDrive/validation"
```

```python
from tensorflow.keras.preprocessing import image_dataset_from_directory
data_train = image_dataset_from_directory(
    data_train_path,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(180, 180),
    batch_size=32)
```

```python
data_cat=data_train.class_names
data_cat
```

```python
data_test=tf.keras.utils.image_dataset_from_directory(data_test_path,shuffle=True,image_size=(256,256),batch_size=32,validation_split=False)
```

```python
from tensorflow.keras.preprocessing import image_dataset_from_directory
data_val = image_dataset_from_directory(
    data_train_path,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(180, 180),
    batch_size=32)
```

```python
plt.figure(figsize=(10,10))
for images,labels in data_train.take(1):
  for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(data_cat[labels[i]])
    plt.axis("off")
```

```python
model=keras.Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(16,3,padding="same",activation="relu"),
    layers.MaxPooling2D(),
    layers.Conv2D(32,3,padding="same",activation="relu"),
    layers.MaxPooling2D(),
    layers.Conv2D(64,3,padding="same",activation="relu"),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dropout(0.2),
    layers.Dense(128),
    layers.Dense(len(data_cat))

    ])
```

```python
data_train
```

```python
model.compile(optimizer="adam",loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=["accuracy"])
```

```python
from re import VERBOSE
epochs=10
history=model.fit(data_train,validation_data=data_val,epochs=epochs)
```

```python
epochs_range=range(epochs)
plt.figure(figsize=(8,8))
plt.subplot(1,2,2)
plt.plot(epochs_range,history.history["accuracy"],label="Training Accuracy")
plt.plot(epochs_range,history.history["val_accuracy"],label="Validation Accuracy")
# plt.legend(loc="lower right")
plt.title("Accuracy")



# plt.subplot(1,2,2)
# plt.plot(epochs_range,history.history["loss"],label="Training Loss")
# plt.plot(epochs_range,history.history["val_loss"],label="Validation Loss")
# # plt.legend(loc="lower right")
# plt.title("Loss")
```

```python
plt.subplot(1,2,2)
plt.plot(epochs_range,history.history["loss"],label="Training Loss")
plt.plot(epochs_range,history.history["val_loss"],label="Validation Loss")
# plt.legend(loc="lower right")
plt.title("Loss")
```

```python
img = tf.keras.preprocessing.image.load_img("/content/drive/MyDrive/test/electric car/electric__image-102.jpeg", target_size=(180, 180))
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0)
```

```python
predict=model.predict(img_array)
```

```python
score=tf.nn.softmax(predict)
```

```python
print("It is image of {} with accuracy score of {:0.2f}".format(data_cat[np.argmax(score)],np.max(score)*100))
```
